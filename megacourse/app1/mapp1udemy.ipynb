```python
map = folium.Map(location=[45,60],tiles = "Stamen Terrain",zoom_start = 2,)#some attributes zoom_start = 5 
```

```python
fg = folium.FeatureGroup(name ="mapping" )
```

```python
df = pd.read_csv('Volcanoes.txt')
```

```python
def color_gen(elevator):
  if elevator >3500:
    color = 'red'
  if elevator >1500:
    color = "orange"
  else:
    color = 'blue'
  return color
```

```python
df.head()
```

```python
for t in range(len(df)) :
  i,j, y  = df.iloc[t,8], df.iloc[t,9], df.iloc[t,5]
  fg.add_child(folium.CircleMarker(location = [i,j], popup=str(y)+" m",  radius=5, fill_color= color_gen(df.iloc[t,5]),color = 'gray',fill_opacity = 0.6))
map.add_child(fg)
```

```python
fg.add_child(folium.Marker(location = [34,45],popup='hi man', icon = folium.Icon(color='green')))
map.add_child(fg)
```

```python
map.save('Map1.html')
```
